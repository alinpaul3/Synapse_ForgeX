# 02 - Feature Checks

This notebook verifies feature engineering and data quality.

```python
import pandas as pd
import numpy as np
from pathlib import Path
import sys

# Add src to path
sys.path.insert(0, str(Path.cwd().parent))

from src.features.text_vectorizer import TextVectorizer
from src.features.metadata_features import MetadataFeatures
from src.data_loader.validate_schema import SchemaValidator

print('Dependencies imported successfully')
```

## Text Vectorization

```python
# Load sample texts
sample_texts = [
    "I love this product, it works great",
    "This is terrible, I hate it",
    "Amazing quality and great service",
]

# Initialize and fit vectorizer
vectorizer = TextVectorizer(max_features=50)
vectors = vectorizer.fit_transform(sample_texts)

print(f'Vector shape: {vectors.shape}')
print(f'Sample vector:\n{vectors[0][:10]}')
```

## Metadata Features

```python
# Create sample data
sample_data = pd.DataFrame({
    'text': sample_texts,
    'age': [25, 35, 45],
    'gender': ['M', 'F', 'M'],
})

# Extract metadata features
metadata_features = MetadataFeatures.extract(sample_data)

print(f'Metadata features shape: {metadata_features.shape}')
print(f'Features:\n{metadata_features}')
```

## Schema Validation

```python
# Define schema
validator = SchemaValidator(
    required_columns=['text', 'age', 'gender'],
    column_types={'age': 'int', 'gender': 'str'},
    min_rows=1,
)

# Validate data
is_valid = validator.validate(sample_data)
print(f'Data is valid: {is_valid}')
```
